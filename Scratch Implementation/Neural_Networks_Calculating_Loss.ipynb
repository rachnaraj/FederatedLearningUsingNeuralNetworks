#### Softmax Function
*   Input --> Exponentiation (e^X) --> Normalisation (e^xi/sum(e^j)) --> Output
*   softmax is actually the combination of [Exponentiation and Normalization]
*   With softmax, we are actually not giving actual values, instead probabilities for each class
softmax.JPG


#### Loss function for Classification Problems with Softmax function as output layer activation function is **Categorical_cross_entropy**

* With Categorical cross entropy, when confidence is higher for the correct class, loss is lower
* But at the same time, if confidence is lower for the correct class, the loss will be higher

**the above two points are precisely what we want**

### One-Hot vector to represent outputs:

```
Total classes - 3

label - 0
one-hot: [1, 0, 0]

label-1
one-hot: [0, 1, 0]

label-2
one-hot: [0, 0, 1]
```



````python
import math

# loss function's output for higher confidence on correct level
softmax_output = [0.7, 0.1, 0.2] # any assumed final output for a NN
target_output = [1, 0, 0] # one-hot encoded labeling

loss_07 = -(target_output[0] * math.log(softmax_output[0])  +
         target_output[1] * math.log(softmax_output[1]) +
         target_output[2] * math.log(softmax_output[2]) )

print("loss_07 = ", loss_07)

# loss function's output for a bit confidence on correct level
softmax_output = [0.3, 0.1, 0.2] # any assumed final output for a NN
target_output = [1, 0, 0] # one-hot encoded labeling

loss_03 = -(target_output[0] * math.log(softmax_output[0])  +
         target_output[1] * math.log(softmax_output[1]) +
         target_output[2] * math.log(softmax_output[2]) )

print("loss_03 = ", loss_03)
````

### Observation: Loss value is more, when confidence is less for correct label
